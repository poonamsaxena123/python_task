```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, np.nan, 35],
    'Address': ['NY', np.nan, 'Chicago'],
    'Salary': [50000, np.nan, 70000]
}
df = pd.DataFrame(data)
new={
    "Age":34,
    "Address":"unknown",
    "Salary":100
}
d=df.fillna(value=new)
d
d['Salary'] = d['Salary'].astype(int)
d.info()
```

```python
fill_values = {
    'Name': 'Unknown',         # String column
    'Age': 56,                  # Numeric column
    'Address': 'Unknown',      # String column
    'Salary': 0.0              # Numeric column
}
df = df.fillna(value=fill_values)
```
