```python
# Divide the columns into numeric columns and categorical columns, then use the fillna method to fill numeric columns with -999, fill categorical columns with 'UNKNOWN'
def fill_null_values(col):
    if col.dtype=="int":
        return col.fillna(0)
    else:
        return col.fillna("unknown")
dist=dist.apply(fill_null_values)
dist
```

```python
# only for cat col 
cat_col=trans.select_dtypes(include="object").columns

trans[cat_col]=trans[cat_col].fillna("unknown")
```

```python
# only for cat col 
num_col=trans.select_dtypes(include="number").columns

trans[num_col]=trans[num_col].fillna(method="ffill")
```
