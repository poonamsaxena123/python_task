```python
# find the top 3 goners  total earning 
df.groupby("Genre")["Gross"].sum().sort_values(ascending=False).head(3)
```

```python
# find the geners whos highest imd rating 
df.groupby("Genre")["IMDB_Rating"].mean().sort_values(ascending=False).head(1)
```

```python
# find the directore name whos papular 
df.groupby("Director")["No_of_Votes"].sum().sort_values(ascending=False).head(3).reset_index()
```

```python
# find the highest rated movie of each geners 
df.groupby(["Genre","Series_Title"])["IMDB_Rating"].sum().sort_values(ascending=False).head(1)
```

```python
# find the movie done y each actore 
df.groupby("Star1")["Series_Title"].count().sort_values(ascending=False)
```
