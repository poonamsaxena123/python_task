```python
# plot 
ax=sns.countplot(x="Gender",data=df,hue="Gender")
ax
for a in ax.containers:
    ax.bar_label(a)
```

```python
# groupy in Gender and sum of amount 
result=df.groupby("Gender")["Amount"].sum().sort_values(ascending=False).reset_index()
```

```python
# female are shoping more according to man
sns.barplot(x="Gender",y="Amount",data=result,hue="Gender")
plt.show()
```

```python
#  which age gruop most purchase product
sns.countplot(x='Age Group',hue="Gender",data=df)
plt.show()
```

```python
result_of_age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
result_of_age
```

```python
# we can see that most of the shoppping are of age  group 26-35 year that female
sns.barplot(x="Age Group",y="Amount",data=result_of_age,hue="Age Group")
plt.show()
```

```python
#total numer of orders from top of the state top 10
order_state=df.groupby("State")["Orders"].sum().sort_values(ascending=False).head(10).reset_index()
order_state
```

```python
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x="State",y="Orders",data=order_state)
plt.show()
```

```python
# total amount from top 10n state
total_amount=df.groupby("State")["Amount"].sum().sort_values(ascending=False).head(10).reset_index()
total_amount
```

```python
# graph of that result
plt.figure(figsize=(10,5))
sns.barplot(x="State",y="Amount",data=total_amount)
plt.xticks(rotation=15)
plt.show()
```

```python
# we can see that most of the uyer  are working in it Avition and Healthcare sectore
occupation_data=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)

occupation_data
plt.figure(figsize=(15,5))
AX =sns.barplot(x="Occupation",y="Amount",data=occupation_data)
```

```python
# we can see that most of the uyer  are working in it Avition and Healthcare sectore
plt.figure(figsize=(15,5))
AX =sns.countplot(x="Occupation",data=df)
for a in AX.containers:
    AX.bar_label(a)
```

```python
# product category
product_data=df.groupby("Product_Category")["Amount"].sum().sort_values(ascending=False).reset_index()
product_data
plt.figure(figsize=(15,5))
ax=sns.barplot(x="Product_Category",y="Amount",data=product_data)
plt.xticks(rotation=20)
plt.show()
```

```python
ax=sns.countplot(x="Product_Category",data=df)
for i in ax.containers:
    ax.bar_label(i)

plt.xticks(rotation=20)
plt.show()
```
