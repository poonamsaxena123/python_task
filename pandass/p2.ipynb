```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/home/developer/Downloads/iris.csv")
```

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data['SepalLengthCm'],data["PetalLengthCm"],c=data["Species"],cmap="summer")
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
data["Species"]=data["Species"].replace({"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2})
plt.colorbar()
plt.text(4.5,1,'point-1')
plt.show()
```

```python
data.sample()
```

```python
#suplot
fig,ax=plt.subplots()
ax.plot(data["SepalLengthCm"],data["PetalLengthCm"])
ax.set_title("mapping")
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(data["SepalLengthCm"],data["PetalLengthCm"],color='red')
ax[1].scatter(data["SepalLengthCm"],data["SepalWidthCm"])
ax[0].set_title("SepalLength and PatelLenghth")
ax[1].set_title("SepalLength and sepalWidth")
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y)
axs[0, 1].scatter(x, y)
axs[1, 0].bar(x, y)
axs[1, 1].hist(y, bins=5)

plt.show()
```

```python
# 3D plot using mat
data2=pd.read_csv("/home/developer/Downloads/batter (1).csv")
```

```python
# 3D scatter plot
fig=plt.figure()
ax=plt.subplot(projection='3d')
ax.scatter3D(data2["runs"],data2["avg"],data2["strike_rate"])
ax.set_title("IPL Analysis")
ax.set_xlabel("runs")
ax.set_ylabel("avg")
ax.set_zlabel("strike_rate")
plt.show()
```

```python
# 3D line plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
z=[10,3,4,56,89]
fig= plt.figure()
ax=plt.subplot(projection="3d")
ax.plot3D(x,y,z,color='red')
```

```python
import numpy as np
# surface plotsx
x=np.linspace(-10,0,100)
y=np.linspace(-10,10,100)
xx ,yy=np.meshgrid(x,y)
xx.shape
```
