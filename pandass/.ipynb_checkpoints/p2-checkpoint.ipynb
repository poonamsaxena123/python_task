```python
import numpy as np
# surface plotsx
x=np.linspace(-10,0,100)
y=np.linspace(-10,10,100)
xx ,yy=np.meshgrid(x,y)
xx.shape
```

```python
z= xx**2 + yy **2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap="viridis")                                   
fig.colorbar(p)
plt.show()
```

```python
fig=plt.figure(figsize=(10,8))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap="viridis")                                   
fig.colorbar(p)
plt.show()
```

```python
fig=plt.figure(figsize=(6,4))
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap="viridis")                                   
fig.colorbar(p)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [200, 300, 550, 760, 1100]
```

```python
# 1.line plot
plt.plot(x,y)
plt.title("simple line plot")
plt.xlabel("score")
plt.ylabel("num")
plt.show()
```

```python
# 2 scatter plot
plt.scatter(x,y)
plt.show()
```

```python
# 3 bar plot
plt.bar(x,y)
plt.show()
```

```python
# 4 histogram plot
plt.hist(x,bins=7)
plt.show()
```

```python
# 5 pie plot
plt.pie(x,labels=y)
plt.show()
```

```python
# 6 sub plot


# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# First subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('First Plot')

# Second subplot
axs[0, 1].scatter(x, y)
axs[0, 1].set_title('Second Plot')

# Third subplot
axs[1, 0].bar(x, y)
axs[1, 0].set_title('Third Plot')

# Fourth subplot
axs[1, 1].hist(data, bins=4)
axs[1, 1].set_title('Fourth Plot')

# Adjust layout
plt.tight_layout()
plt.show()
plt.savefig("plot.png")
```

```python
import pandas as pd

# Sample Data
data = {
    'Restaurant Name': ['The Gourmet Kitchen', 'Seafood Delight', 'Pasta Palace', 'Spicy Corner', 'Sweet Treats', 
                        'Vegan Delights', 'Steakhouse Supreme', 'Sushi World', 'Burger Haven', 'Pizza Paradise'],
    'Cuisine': ['Continental', 'Seafood', 'Italian', 'Indian', 'Desserts', 'Vegan', 'Steakhouse', 'Japanese', 'Fast Food', 'Italian'],
    'Location': ['Downtown', 'Waterfront', 'City Center', 'Suburb', 'City Center', 'Downtown', 'Suburb', 'Downtown', 'City Center', 'Waterfront'],
    'Average Cost for Two': [50, 70, 40, 30, 20, 35, 80, 60, 25, 45],
    'Rating': [4.5, 4.7, 4.3, 4.1, 4.8, 4.2, 4.6, 4.9, 4.0, 4.4]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Sample Restaurant Dataset:")
print(df)
```
