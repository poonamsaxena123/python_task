```python
# to find outlier of open col 
q1=df["Open"].quantile(0.25)
q2=df["Open"].quantile(0.75)
```

```python
iqr=q2-q1
```

```python
iqr
```

```python
up=q2+1.5*iqr
up
```

```python
lo=q1-1.5*iqr
lo
```

```python
# find outliers

out=df[(df["Open"]<=lo) | (df["Open"]>=up)]
```

```python
df
```

```python
# remove ouliyers
df=df[(df["Open"]>=lo) & (df["Open"]<=up)]
```

```python
# To convert in date time format 
df["Date"]=pd.to_datetime(df["Date"])
```

```python
# Determine the daily returns and plot the histogram of daily returns. 
df["Day"]=df["Date"].dt.day
df
df["year"]=df["Date"].dt.year
df
```

```python
#Calculate and plot the moving average of the Close price over a 30-day window.

dff=df["Close"].rolling(window=30).mean()
dff
```

```python

# Plot the Close prices and the 30-day moving average
plt.figure(figsize=(14, 7))
sns.lineplot(x=df['Date'], y=df['Close'],label='Close Price', color='blue')
plt.plot(df['Date'], dff, label='30-Day Moving Average', color='orange')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Close Price and 30-Day Moving Average')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot the Close price over time.  
sns.lineplot(x=df["Date"],y=df["Close"],color="orange")
plt.grid(True)

plt.show()
```

```python
# Plot the Volume of trades over time.
plt.figure(figsize=(10,6))
sns.lineplot(x="Date",y="Volume",data=df,color="green")
plt.grid(True)
plt.show()
```

```python
# Compute the correlation matrix for the columns Open, High, Low, Close, Adj Close, and Volume. 
new_matrix=df[["Open","High","Low","Close","Adj Close","Volume"]].corr()
```
