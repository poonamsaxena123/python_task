```python
import pandas as pd
import numpy as np

# Create a sample weather dataset
data = {
    'day': ['2024-07-01', '2024-07-02', '2024-07-03', '2024-07-04', '2024-07-05', 
            '2024-07-06', '2024-07-07', '2024-07-08', '2024-07-09', '2024-07-10'],
    'event': ['Sunny', 'Rain', np.nan, 'Sunny', 'Snow', 'Rain', 'Sunny', np.nan, 'Rain', 'Sunny'],
    'temperature': [30, 25, 27, np.nan, -5, 22, 28, 30, 21, np.nan],
    'windspeed': [10, np.nan, 7, 8, 5, 10, np.nan, 6, 8, 7],
    'city': ['New York', 'Los Angeles', 'New York', 'Houston', 'Phoenix', 
             'New York', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']
}

# Create the DataFrame
dff = pd.DataFrame(data)

# Display the DataFrame

dff
```
